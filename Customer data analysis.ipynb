```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = dataset['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linmodel = LinearRegression()
linmodel.fit(X_train,y_train)
```

```python
predictions = linmodel.predict(X_test)
```

```python
predictions
```

```python
linmodel.predict([[33,13,36,4]])
```

```python
linmodel.score(X_test,predictions)
```

```python
import sklearn.metrics as metrics
```

```python
print('MAE: {}'.format(metrics.mean_absolute_error(y_test, predictions)))
```

```python
print('MSE: {}'.format(metrics.mean_squared_error(y_test, predictions)))
```

```python
print('RMSE: {}'.format(np.sqrt(metrics.mean_squared_error(y_test, predictions))))
```

```python
pd.DataFrame(linmodel.coef_ , X.columns, columns=['Coeffecient'])
```
